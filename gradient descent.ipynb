```python
path = r"C:\Users\Dell\Desktop\PRSA_data_2010.1.1-2014.12.31.csv"
f = open(path, 'r')
```

```python
dataset = []
header = f.readline().strip().split(',')
for line in f:
    line = line.split(',')
    dataset.append(line)
```

```python
header.index('pm2.5')
```

```python
dataset = [d for d in dataset if d[5] != 'NA']
```

```python
dataset = dataset[:1000]
```

```python
#extract features from dataset
```

```python
X = [[1,float(d[7])] for d in dataset]
y = [float(d[5]) for d in dataset]
```

```python
X[0]
```

```python
K = len(X[0])
K
```

```python
#initialize parameter
theta = [0.0] * K
```

```python
#initialize theta_0 to the mean value will help to converge faster
theta[0] = sum(y) / len(y)
theta[0]
```

```python
def inner(x,y):
    return sum([a*b for (a,b) in zip(x,y)])
```

```python
def norm(x):
    return sum([a*a for a in x])
```

```python
def derivative(X,y,theta):
    dtheta = [0.0] * len(theta)
    K = len(theta)
    N = len(X)
    MSE = 0
    for i in range(N):
        error = inner(X[i], theta) - y[i]
        for k in range(K):
            dtheta[k] += 2 * X[i][k] * error/N
        MSE += error * error / N
    return dtheta, MSE
```

```python
learning_rate = 0.0003
```

```python
MM = [0]
```

```python
while(True):
    dtheta, MSE = derivative(X,y,theta)
    MM.append(MSE)
    print(MSE)
    for k in range(K):
        theta[k] -= learning_rate * dtheta[k]
    
    if abs(MSE - MM[-2]) < 0.0001:break
    
```
